```python
import numpy as np
import pandas as pd
Network_Traffic = np.load('Network_Traffic.npy', allow_pickle=True).item()
Network_Energy = np.load('Network_Energy.npy', allow_pickle=True).item()
Network_Energy_std = np.load('Network_Energy_std.npy', allow_pickle=True).item()
Network_Carbon = np.load('Network_Carbon.npy', allow_pickle=True).item()
Network_Carbon_std = np.load('Network_Carbon_std.npy', allow_pickle=True).item()
Additional_Carbon = np.load('Additional_Carbon_Province.npy', allow_pickle=True).item()
Additional_Carbon_std = np.load('Additional_Carbon_std_Province.npy', allow_pickle=True).item()
Wasted_Energy = np.load('Wasted_Energy_Province.npy', allow_pickle=True).item()
Wasted_Energy_std = np.load('Wasted_Energy_std_Province.npy', allow_pickle=True).item()
Time_list_After5G = []
for year in ['2021', '2022', '2023']:
    for month in range(1, 13):
        if month < 10:
            Time_list_After5G.append(f'{year}-0{month}')
        else:
            Time_list_After5G.append(f'{year}-{month}')
len(Time_list_After5G), len(Network_Traffic['Beijing'])
```

```python
for p in Network_Traffic:
    law = p
    for i in range(24, 36):
        law += f' & {round(Network_Traffic[p][i]*30, 2)}$\pm${round(Network_Traffic[p][i]*30*0.03, 2)}'
    law += '\\\\ \n\hline'
    print(law)
```

```python
Energy_all = {'2021':0, '2022':0, '2023':0}
Energy_all_std = {'2021':0, '2022':0, '2023':0}
for p in Network_Energy:
    law = p
    for i in range(3):
        value = 0
        value_std = 0
        for j in range(12):
            value += Network_Energy[p][i*12+j]*30
            value_std += Network_Energy_std[p][i*12+j]*30
        law += f' & {round(value/1000, 2)}$\pm${round(value_std/1000, 2)}'
        Energy_all['202'+str(i+1)] += value
        Energy_all_std['202'+str(i+1)] += value_std
    law += '\\\\ \n\hline'
    print(law)
print(f"All of China & {round(Energy_all['2021']/1000, 2)}$\pm${round(Energy_all_std['2021']/1000, 2)} & {round(Energy_all['2022']/1000, 2)}$\pm${round(Energy_all_std['2022']/1000, 2)} & {round(Energy_all['2023']/1000, 2)}$\pm${round(Energy_all_std['2023']/1000, 2)}")
```

```python
Carbon_all = {'2021':0, '2022':0, '2023':0}
Carbon_all_std = {'2021':0, '2022':0, '2023':0}
for p in Network_Carbon:
    law = p
    for i in range(3):
        value = 0
        value_std = 0
        for j in range(12):
            value += Network_Carbon[p][i*12+j]*30
            value_std += Network_Carbon_std[p][i*12+j]*30
        law += f' & {round(value/1000, 2)}$\pm${round(value_std/1000, 2)}'
        Carbon_all['202'+str(i+1)] += value
        Carbon_all_std['202'+str(i+1)] += value_std
    law += '\\\\ \n\hline'
    print(law)
print(f"All of China & {round(Carbon_all['2021']/1000, 2)}$\pm${round(Carbon_all_std['2021']/1000, 2)} & {round(Carbon_all['2022']/1000, 2)}$\pm${round(Carbon_all_std['2022']/1000, 2)} & {round(Carbon_all['2023']/1000, 2)}$\pm${round(Carbon_all_std['2023']/1000, 2)}")
```
